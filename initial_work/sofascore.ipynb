```python
import ScraperFC as sfc
```

```python
MATCH_ID = 14566935
ss = sfc.Sofascore()
```

```python
players = ss.scrape_player_match_stats(MATCH_ID)
players.columns
```

```python
players = players.loc[:, ~players.columns.duplicated()]
players = players[players["position"] != "G"]
players = players[players["substitute"] == False]
```

```python
players = players[["position","teamName","name", "minutesPlayed", "duelWon", "duelLost"]]
players = players.fillna(0.0)
players["duelTotal"] = players["duelWon"] + players["duelLost"]
players
```

```python
players = players.sort_values(by=["duelTotal", "duelWon"], ascending=[False, False])
gala_players = players[players["teamName"] == "Galatasaray"]
gala_players
```

```python
teams = ss.get_team_names(MATCH_ID)
for team in teams:
    if team != "Galatasaray":
        opponent = team
opponent_players = players[players["teamName"] == opponent]
opponent_players
```

```python
game = ss.scrape_team_match_stats(MATCH_ID)
game = game[game["period"] == "ALL"]
percentage = game[game["name"] == "Duels"]
ground = game[game["name"] == "Ground duels"]
aerial = game[game["name"] == "Aerial duels"]
```

```python
if teams[0] == "Galatasaray":
    home = "Galatasaray"
else:
    home = opponent
```

```python
gala = {}
oppo = {}
if home == "Galatasaray":
    gala["percentage"] = percentage["home"]
    oppo["percentage"] = percentage["away"]
    gala["ground"] = ground["homeValue"]
    oppo["ground"] = ground["awayValue"]
    gala["aerial"] = aerial["homeValue"]
    oppo["aerial"] = aerial["awayValue"]
else:
    gala["percentage"] = percentage["away"]
    oppo["percentage"] = percentage["home"]
    gala["ground"] = ground["awayValue"]
    oppo["ground"] = ground["homeValue"]
    gala["aerial"] = aerial["awayValue"]
    oppo["aerial"] = aerial["homeValue"]

# gala["total"] = gala["ground"] + gala["aerial"]
# oppo["total"] = oppo["ground"] + oppo["aerial"]
```
